```python
def string_match_brute(palabra):
  r=0
  archivo = open("el_quijote_ii.txt", "r")
  lineas=archivo.read()
  for i in range ( len(lineas) - len(palabra) + 1 ):
    for j in range (len(palabra)):
      if palabra[j] == lineas[i+j] :
        pass 
      else:
        break
    if j+1 == len(palabra) and lineas[i+j] == palabra[j]:
      r+=1
      print(f"Se encontró en la posicion: {i}")
  print(f"Las veces que {palabra} se encuentra es: {r}")

ejemplo = "molino"
string_match_brute(ejemplo)
print("----------------------")
ejemplo = "hidalgo"
string_match_brute(ejemplo)
print("----------------------")
ejemplo = "merced"
string_match_brute(ejemplo)
print("----------------------")

```
